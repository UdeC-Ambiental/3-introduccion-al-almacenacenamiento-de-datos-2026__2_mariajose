# ⚙️ Reto 2 – Filtros y condiciones en Pandas

**Objetivo:** aplicar filtros y condiciones lógicas para clasificar los datos.

```python
import pandas as pd
datos = pd.read_csv('../data/4_resultados_agua.csv')
```

### 🔹 Ejemplo de comparación
Puedes filtrar las filas según una condición, por ejemplo pH mayor a 8.5:

```python
datos[datos['pH'] > 7.1]
```

### 💡 Actividad
- Crea al menos dos condiciones (por pH y temperatura).
- Genera una nueva columna con la categoría resultante.

```python
# Condición 1: puntos con pH menor a 7
datos[datos["pH"] < 7]
```

```python
# Condición 2: puntos con temperatura mayor o igual a 24
datos[datos["Temp"] >= 24]
```

```python
datos["estado_pH"] = "aceptable"
datos.loc[datos["pH"] < 7, "estado_pH"] = "bajo"
```

```python
datos["categoria"] = "pH aceptable y temperatura alta"

datos.loc[(datos["pH"] < 7) & (datos["Temp"] < 22), "categoria"] = "pH bajo y temperatura baja"

datos.loc[(datos["pH"] < 7) & (datos["Temp"] >= 22), "categoria"] = "pH bajo y temperatura alta"

datos.loc[(datos["pH"] >= 7) & (datos["Temp"] < 24), "categoria"] = "pH aceptable y temperatura moderada"
```

```python
datos["categoria"].value_counts()
```

## Conclusión

En este notebook se aplicaron filtros sobre las variables `pH` y `Temp` para seleccionar registros específicos del dataset.  
Además, se creó una nueva columna llamada `Categoria`, que clasifica cada punto según sus valores de pH y temperatura.

Esto permite organizar mejor los datos y facilita su interpretación en análisis posteriores.
