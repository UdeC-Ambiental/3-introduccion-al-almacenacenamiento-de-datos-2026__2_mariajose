## 🧠 Reflexión
Describe con tus palabras qué ventajas tiene usar `pandas` frente a listas o diccionarios para manejar datos tabulares.

`pandas` facilita mucho el trabajo con datos tabulares porque organiza la información en filas y columnas, de forma similar a una tabla de Excel. A diferencia de listas o diccionarios, permite filtrar, resumir y analizar datos de manera más rápida y ordenada. Además, incluye funciones como `head()`, `info()` y `describe()`, que ayudan a explorar los datos sin tener que programar todo manualmente.
